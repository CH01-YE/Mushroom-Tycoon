```python
import os
import json
import pandas as pd
```

```python
path = './mushroom_data/data/1.Training/label_data/TL5_shiitake/Blight'
```

```python
data=os.listdir(path) # 폴더 안에 있는 json 파일
```

```python
data
```

```python
len(data)
```

```python
meta_data = []
name_data = []

for i in range(len(data)):
    with open(f'./mushroom_data/data/1.Training/label_data/TL5_shiitake/Blight/{data[i]}','r', encoding='UTF-8') as f:
        json_data = json.load(f)
        
    meta_data.append(json_data['META'])
    name_data.append(json_data['INFO']['CATEGORY_NAME'])
```

```python
meta_data
```

```python
name_data
```

```python
test=pd.DataFrame(meta_data)
```

```python
test
```

```python
test['NAME'] = name_data
```

```python
test
```

```python
test.to_csv('shiitake_Blight.csv')
```
